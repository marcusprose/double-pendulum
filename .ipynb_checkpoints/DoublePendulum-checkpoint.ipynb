## Motivation

The angle of a single pendulum follows a pretty straight forward motion when graphed over time (close to a sine wave). However, a double pendulum introduces the idea of chaotic motion, which makes it much harder to estimate the angle. The goal of this project is to try to use a couple of numerical methods in order to estimate the angle theta over time while obseriving some interesting phenomena.

## Lagrangian: Finding the Equations for $\dot \theta_1$ and  $\dot \theta_2$ 

The Lagrangian is the ideal method for this. The goal is to find the difference between potential and kinetic energy. That is, to minimize L where 

\begin{equation}
    L = KE - PE 
\end{equation}


by solving the ordinary differential equation 



$$ \frac{\partial L}{\partial q_i} - \frac{d}{dt} ( \frac{L}{\partial \dot q_i} ) = 0, $$

where $\dot q_i = \frac{d q_i}{dt}$ and $\ddot q_i = \frac{d^2 q_i}{dt^2}$. 

Furthermore, the following equations should be established,

\begin{equation}
\begin{split}
KE &= \frac{1}{2} m_1 (\dot x_1^2 + \dot y_1^2) + \frac{1}{2} m_2 (\dot x_2^2 + \dot y_2^2) \\
 PE &= m_1gh_1 + m_1gh_2,
\end{split}
\end{equation}




\begin{align}
    x_1 = L_1 \sin\theta_1 && 
    x_2 &= L_2 \sin\theta_2  L_1\sin\theta_1 \\ 
    y_1 = -L1\cos\theta_1 &&
    y_2 &= -L_2 \cos\theta_2 - L_1\cos\theta_1
\end{align}




Now, note the time derivatives of $x_1$ and $y_1$, the kinetic energy can be derived. 

$$ \dot x_1 = L_1\cos\theta_1\dot\theta_1 $$
$$ \dot y_1 = L_1\sin\theta_1\dot\theta_1 $$

\begin{equation} 
\begin{split}
KE_1 &= \frac{1}{2} m_1 (\dot x_1^2 + \dot y_1^2) = \frac{1}{2} m_1 [(L_1\cos\theta_1\dot\theta_1)^2 + (L_1 \sin\theta_1\dot\theta_1)^2 ] \\
  &= \frac{1}{2} m_1 [L_1^2 \sin^2\theta_1\dot\theta_1^2 + L_1^2\cos^2\theta_1\dot\theta_1^2] \\
  &= \frac{1}{2} m_1 L_1^2 \theta_1^2 (\cos^2 \theta_1 + \sin^2 \theta_1) \\
  &= \frac{1}{2} m_1 L_1^2 \dot \theta_1^2.
\end{split}
\end{equation}

Now, let's move to potential energy, which is a little simpler: 

$$ U_1 = -m_1 g L_1 \cos\theta_1.$$

This completes the general derivation for the first body of motion. Now, let's compute the time derivatives for the second mass of the pendulum. 


$$ x_2 = L_2 \sin\theta_2 + L_1 \sin\theta_1 \implies \dot x_2 = L_2 \cos\theta_2 \dot\theta_2 + L_1 \cos\theta_1 \dot\theta_1 $$

$$ y_2 = -L_2 \cos\theta_2 - L_1 \cos\theta_1 \implies \dot y_2 = L_2 \sin\theta_2 \dot\theta_2 + L_1 \sin\theta_1 \dot\theta_1$$

Then, kinetic energy for the second angle is as follows, 

\begin{equation}

 &= \frac{1}{2} m_2 [(L_2 \cos\theta_2 \dot\theta_2 + L_1 \cos\theta_1 \dot\theta_1)^2 + (L_2 \sin\theta_2 \dot\theta_2 + L_1\sin\theta_1 \dot\theta_1)^2 ]
\end{split}
\end{equation}

## Discretization

Again, the derivation yields the equations (slightly rearranged and simplified)

$$ (m_1 + m_2)L_1\ddot\theta_1 + m_2L_2\ddot\theta_2\cos(\theta_1 - \theta_2) + m_2L_2\dot\theta_2\sin(\theta_1 - \theta_2) + (m_1 + m_2)g\sin\theta_1 = 0 $$

$$
m_2L_2\ddot\theta_2 + m_2L_1\ddot\theta_1\cos(\theta_1 - \theta_2) - m_2L_1\dot\theta_1^2\sin(\theta_1 - \theta_2) + m_2g\sin(\theta_2) = 0 $$ 

$$\implies
L_2\ddot\theta_2 + L_1\ddot\theta_1\cos(\theta_1 - \theta_2) - L_1\dot\theta_1^2\sin(\theta_1 - \theta_2) + g\sin(\theta_2) = 0 $$

Now, establish the following relationship: from $A\textbf{x} = \textbf{b}$,

$$ M\vec{\ddot\theta} = \vec{F} $$

$$ \begin{bmatrix} (m_1+m_2)L_1 && m_2 L_2 \cos(\theta_1 - \theta_2) \\
       L_1\cos(\theta_1 - \theta_2) && L_2 \end{bmatrix} \begin{bmatrix} \ddot \theta_1 \\ \ddot \theta_2 \end{bmatrix} = \begin{bmatrix} -m_2L_2\dot\theta_2^2 \sin(\theta_1 - \theta_2) - (m_1 + m_2)g\sin\theta_1 \\ L_1\dot\theta_1^2 \sin(\theta_1 - \theta_2) - g\sin\theta_2
       \end{bmatrix}$$
       
To solve for the time derivatives, we need to invert the mass matrix. That is,

$$ \vec{\ddot \theta} = M^{-1}\vec{F}. $$

To put our solution into a more usable format, we put our solution into state space form, which we'll call SSF (reference: https://en.wikipedia.org/wiki/State-space_representation). So,

$$ \vec{y} = \begin{bmatrix} \dot\theta_1 \\ \dot\theta_2 \end{bmatrix} \implies \vec{\dot y} = \frac{d}{dx} \begin{bmatrix} \dot\theta_1 \\ \dot\theta_2 \end{bmatrix} = \frac{d}{dx} \begin{bmatrix} M^{-1} \vec{F} \\ \vec{\dot\theta} \end{bmatrix} = \vec{SSF} (\vec{y}, t) $$

The Runge Kutta method iteration is as follows,

$$ \vec{y_{t + 1}} = \vec{y_t} + \delta t \cdot \vec{SSF} $$

## Modeling $\theta_1$ and $\theta_2$ w.r.t. time
We can now use the above Lagrangian derivation to model the angle movements using a few numerical methods. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.integrate import odeint
from numpy.linalg import inv
import time

# magic
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
# a class to implement numerical methods of a double pendulum
#
#
#
class DoublePendulum:
    
    # constructor 
    #
    #
    #
    def __init__(self, g, L1, L2, M1, M2, y, dt, t):
        self.g = g
        self.L1 = L1
        self.L2 = L2
        self.M1 = M1
        self.M2 = M2
        
        self.y = y
        self.dt = dt
        self.t = t # linspace or arange spanning the iteration
    
    
    # state space form  
    #
    # returns an array of the form arr = [theta1', theta2', theta1, theta2]
    # my understanding of this came from Wikipedia: https://en.wikipedia.org/wiki/State-space_representation
    # ** note that they give a great derivation for a single pendulum, the double pendulum follows naturally
    #
    def SSF(self, y, t):
            
            # y is of the form [theta1', theta2', theta1, theta2]
            # what we are solving for
            theta1_dot = y[0]
            theta2_dot = y[1]
            
            # also what we are solving for
            theta1 = y[2] 
            theta2 = y[3]
            
            # masses as described and Mass matrix
            m11 = (self.M1 + self.M2) * self.L1
            m12 = self.M2 * self.L2*np.cos(theta1 - theta2)
            m21 = self.L1 * np.cos(theta1 - theta2)
            m22 =  self.L2
            
            M = np.array([[m11, m12], 
                          [m21, m22]]) 
            
            
            # forces as described and the Force matrix
            f1 = (-self.M2 * self.L2 * (theta2_dot **2 ) * np.sin(theta1 - theta2)) - (self.M1 + self.M2) * self.g * np.sin(theta1)
            f2 = self.L1 * (theta1_dot ** 2) * np.sin(theta1 - theta2) - self.g * np.sin(theta2)
            F = np.array([f1, f2])
            
            # remember, the theta'' matrix needs to be inverted
            a = inv(M).dot(F)

            # returns an array of the form [theta_1'', theta2'', theta1', theta2'] (just accelerations and vel, resp.)
            return np.array([a[0], a[1], theta1_dot, theta2_dot])
        
        
    # numerical method: RK4
    #
    # the equations for the RK4 method were obtained from Wikipedia: 
    # https://en.wikipedia.org/wiki/Runge%E2%80%93Kutta_methods
    #
    # dt = step size
    # y = our unknown vector function at time t
    def RK4(self, y, t, dt):
        k1 = self.SSF(self.y, t)
        k2 = self.SSF(self.y + k1 * self.dt / 2, t + self.dt / 2)
        k3 = self.SSF(self.y + k2 * self.dt / 2, t + self.dt / 2)
        k4 = self.SSF(self.y + k3  * self.dt, t + self.dt)
        return (self.dt / 6) * (k1 + 2 * k2 + 2 * k3 + k4)
    
    
    # numerical method: explicit euler
    #
    #
    # the equation for euler's method was obtained from my last class in numerical methods
    #
    def explicit_euler(self, y, t, dt):
        return self.dt * self.SSF(self.y, self.t)
        

    # implementation of the euler helper method
    #
    #
    #    
    def discretize_explicit_euler(self):
        
        # arrays to populate 
        theta1 = []
        theta2 = []
            
        for i in self.t:
            self.y = self.y + self.explicit_euler(self.y, i, self.dt)
            
            theta1.append(self.y[2])
            theta2.append(self.y[3])
        
        sol = pd.DataFrame([theta1, theta2])
        sol = sol.T
        sol.columns = ['theta1', 'theta2']
        
        return sol
    
    # plot euler
    #
    #
    # 
    def discretize_explicit_euler_plot(self, sol):
        # plot the angles w.r.t time t
        plt.plot(self.t, sol['theta1'])
        plt.plot(self.t, sol['theta2'])
        plt.grid(True)
        plt.title('Pendulum Angles as a function of Time (method=explicit euler)')
        plt.xlabel('time(/s)')
        plt.ylabel('angle(/radians)')
        plt.legend(['Theta1', 'Theta2'], loc='upper right')
        plt.show()
            
            
    # implementation of the RK4 helper method
    #
    #
    #
    def discretize_RK4(self):

        # arrays to populate 
        theta1 = []
        theta2 = []
        
        # begin iterate
        for i in self.t:
            self.y = self.y + self.RK4(self.y, i, self.dt)

            theta1.append(self.y[2])
            theta2.append(self.y[3])
        
        
        sol = pd.DataFrame([theta1, theta2])
        sol = sol.T
        sol.columns = ['theta1', 'theta2']
        
        return sol
    
    
    # plot RK4
    #
    #
    # 
    #
    def discretize_RK4_plot(self, sol):
        # plot the angles w.r.t time t
        plt.plot(self.t, sol['theta1'])
        plt.plot(self.t, sol['theta2'])
        plt.grid(True)
        plt.title('Pendulum Angles as a function of Time (method=RK4)')
        plt.xlabel('time(/s)')
        plt.ylabel('angle(/radians)')
        plt.legend(['Theta1', 'Theta2'], loc='upper right')
        plt.show()
        
    
    
    # numerical method: scipy's built in odeint
    #
    #
    # perhaps the most accurate of the implemented methods.. used for comparison of num methods
    #
    def scipy_integrate(self):

        sol = odeint(self.SSF, y0=self.y, t=self.t)
        sol = pd.DataFrame(sol, columns=['theta`', 'theta2`', 'theta1', 'theta2'])

        return sol 
    
    
    # plot odeint
    #
    #
    # 
    #
    def scipy_integrate_plot(self, sol):
        plt.plot(self.t, sol['theta1'])
        plt.plot(self.t, sol['theta2'])
        plt.grid(True)
        plt.title(f'Pendulum Angles as a function of time (method=scipy odeint)')
        plt.xlabel('time(/s)')
        plt.ylabel('angle(/radians)')
        plt.legend(['Theta1', 'Theta2'], loc='upper right')
        plt.show()
```

### A Brief Digression: Precision

```python
# 
euler_theta_1 = np.array(dp_true.discretize_explicit_euler()['theta1'])
RK4_theta_1 = np.array(dp_true.discretize_RK4()['theta1'])
scipy_theta_1 = np.array(dp_true.scipy_integrate()[['theta1']])
df_theta1 = pd.DataFrame([euler_theta_1, RK4_theta_1, scipy_theta_1], dtype=float)
df_theta1 = df_theta1.T
df_theta1.columns = ['euler_theta_1', 'RK4_theta_1', 'scipy_theta_1']
df_theta1.insert(loc=3,
              column='precision',
               value=df_theta1['euler_theta_1']-df_theta1['RK4_theta_1'])
display(df_theta1)
plt.plot(df_theta1['precision'])
plt.title('Precision as a function of time')
plt.grid(True)
# plt.plot(y=0)
```

Examining the above, we can see that the R.H.S. is reasonably close to zero. But interestingly, repeated measurements also has some wave-like characteristics. However, running the cell many times causes some pretty noticible changes in each calculation each time the cell is run. This is likely due to the way the numerical method calculated. There exists a degree of chaos in this problem, so this is expected. Now, lets dive into the graphs of $\theta_1$ and $\theta_2$. With more iterations, you might see more waves and oscillations. 

```python
dp_true.discretize_explicit_euler_plot(dp_true.discretize_explicit_euler())
dp_true.discretize_RK4_plot(dp_true.discretize_RK4())
dp_true.scipy_integrate_plot(dp_true.scipy_integrate())
```

As shown, the numerical method used and timestep is a large influencer in how these angles are calculated. For the most part, however, the measurements seem to be correct!

There are a lot of intersting obeservations regarding the angles of the double pendulum. Firstly, we're releasing $\theta_2$ from a position of 1, and letting it interate. It does have some wavelike properties, which is something we would come to expect this type of motion. The roughness introduced from the graph is likely due from the additional mass that is affecting the movement of the angles. Also, if you were to look at the suprememum (resp. infimum) at each subinterval such that the wave makes one full iteration, you can also see some wave like characteristics. A graph of the suprema and infimema closely resembles a cosine wave for $\theta_1$ and a sine wave for $\theta_2$.
